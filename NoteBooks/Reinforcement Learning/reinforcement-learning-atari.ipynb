```python
import gym
from stable_baselines3 import A2C
from stable_baselines3.common.vec_env import VecFrameStack # This one is needed to test on 4 envs at the same time
from stable_baselines3.common.evaluation import evaluate_policy
from stable_baselines3.common.env_util import make_atari_env
import os
```

```python
env = make_atari_env('Breakout-v0',n_envs=4,seed=0) # Helps to make atari env
env = VecFrameStack(env,n_stack=4) # Helps to make few envs together
```

```python
log_path = os.path.join('Training','Logs')
model = A2C('CnnPolicy',env,verbose=1,tensorboard_log=log_path) # Using different model and different Policy. PPO was for Tabular Data, but if we observe image - CNN is better
```

```python
model.learn(total_timesteps=3_000_000)
```

# Saving and reloading

```python
a2c_path = os.path.join('Training', 'Saved Models', 'A2C_model')
model.save(a2c_path)
```

```python
del model 
```

```python
model = A2C.load(a2c_path, env)
```

# Evaluating

```python
env = make_atari_env('Breakout-v0',n_envs=1,seed=0) # we changed our envs to 1 because evaulation happens only with one piece
env = VecFrameStack(env,n_stack=4)
```

```python
evaluate_policy(model,env,n_eval_episodes=10, render=False)
```
